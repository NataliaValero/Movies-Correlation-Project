### Import libraries

```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Configurations of the plots
```

### Read in the data

```python
#Read in the data

df = pd.read_csv(r'C:\Users\Natalia\Desktop\Proyectos spyder\Proyecto 4 Data Analyst\movies.csv')
```

```python
# Let's look at the data

df.head()
```

### Look for missing data

```python
# Let's see if there is any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,percent_missing))
    
```

```python
df.isnull().sum()
```

```python
# Delete rows with any null values

df= df.dropna(how='any', axis=0)
df.shape
```

### Data Cleaning

```python
# Data types for our columns

df.dtypes
```

```python
# Convert float to int

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
# Year correct

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
# Order by gross

df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Look all of the data
# I can scroll over the data

# pd.set_option('display.max_rows', None)
# df
```

```python
# Drop any duplicates

df= df.drop_duplicates()
```

### Finding Correlations in the Data

```python
# Budget High Correlation
# Company high correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()


```

```python
# Correlations
# Plot Budget vs Gross using seaborn

sns.regplot(x= 'budget', y= 'gross', data= df, scatter_kws= {'color':'red'}, line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation

# pearson, kendall, spearman
df.corr(method='pearson') 

# High correlation between budget and gross
```

```python
correlation_matriz = df.corr(method='pearson') 

sns.heatmap(correlation_matriz, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looks at company

df.head()
```

```python
# numerized object columns to category type to do correlations

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
        
```

```python
df_numerized.dtypes
```

```python
correlation_matriz2 = df_numerized.corr(method='pearson') 

sns.heatmap(correlation_matriz2, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat2 = df_numerized.corr()

#unstack method

corr_pairs = correlation_mat2.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
